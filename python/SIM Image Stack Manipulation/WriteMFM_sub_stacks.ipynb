```python
scale = True
scaleZ_by = 3

for kk in range(phases):
    curr_path = filepath[:-2] + str(kk) + '/'
    print(curr_path)
    createFolder = curr_path + writepath
    mode = 0o777
    if(os.path.exists(createFolder) == False):
        os.mkdir(createFolder, mode)
    out_filename = 'focus_subStack_' #don't include .tif suffix

    fileNamePrefix = 'single_centered_point_488nm_point_7_phase_26p5nm_'
    fileNamePostfix = '_angle2_phase_' + str(kk) + '.tif'

    x1 = tifffile.imread(curr_path + fileNamePrefix + 'plane1' + fileNamePostfix)
    x2 = tifffile.imread(curr_path + fileNamePrefix + 'plane2' + fileNamePostfix)
    x3 = tifffile.imread(curr_path + fileNamePrefix + 'plane3' + fileNamePostfix)
    x4 = tifffile.imread(curr_path + fileNamePrefix + 'plane4' + fileNamePostfix)
    x5 = tifffile.imread(curr_path + fileNamePrefix + 'plane5' + fileNamePostfix)
    x6 = tifffile.imread(curr_path + fileNamePrefix + 'plane6' + fileNamePostfix)
    x7 = tifffile.imread(curr_path + fileNamePrefix + 'plane7' + fileNamePostfix)
    x8 = tifffile.imread(curr_path + fileNamePrefix + 'plane8' + fileNamePostfix)
    x9 = tifffile.imread(curr_path + fileNamePrefix + 'plane9' + fileNamePostfix)

    IS_SIM = True
    print(x5.shape)
    print(x5.dtype)
    dims = x5.shape
    print(dims[0])
    print(dims[1])
    print(dims[2])

    a = np.zeros((9, dims[2], dims[1])).astype(x5.dtype)

    for i in range(dims[0]):
        a[0,:,:] = x1[i,:,:]
        a[1,:,:] = x2[i,:,:]
        a[2,:,:] = x3[i,:,:]
        a[3,:,:] = x4[i,:,:]
        a[4,:,:] = x5[i,:,:]
        a[5,:,:] = x6[i,:,:]
        a[6,:,:] = x7[i,:,:]
        a[7,:,:] = x8[i,:,:]
        a[8,:,:] = x9[i,:,:]


        dims = a.shape
        if scale:
            b = np.zeros((dims[0]*scaleZ_by, dims[1], dims[2])).astype(a.dtype) 
            for k in range(dims[1]):
                b[:,k,:] = cv2.resize(a[:,k,:], dsize=(dims[1], dims[0]*scaleZ_by), interpolation=cv2.INTER_CUBIC)
                maxb = np.max(b)

            tifffile.imwrite(curr_path + writepath + out_filename + str(i) + '.tif', b)
        else:
            tifffile.imwrite(curr_path + writepath + out_filename + str(i) + '.tif', a)
        # tifffile.imwrite(curr_path + writepath + out_filename + str(i) + '.tif', a)

    # frame_div = 1



    # if IS_SIM:
    #     frame_div = 15

    # a = np.zeros((9*frame_div,dims[2],dims[1])).astype(np.uint16)

    # for i in range(dims[0]//frame_div):
    #     for j in range(frame_div):
    #         a[0*frame_div+j,:,:] = BL[i*frame_div + j,:,:]
    #         a[1*frame_div+j,:,:] = BM[i*frame_div + j,:,:]
    #         a[2*frame_div+j,:,:] = BR[i*frame_div + j,:,:]
    #         a[3*frame_div+j,:,:] = ML[i*frame_div + j,:,:]
    #         a[4*frame_div+j,:,:] = MM[i*frame_div + j,:,:]
    #         a[5*frame_div+j,:,:] = MR[i*frame_div + j,:,:]
    #         a[6*frame_div+j,:,:] = TL[i*frame_div + j,:,:]
    #         a[7*frame_div+j,:,:] = TM[i*frame_div + j,:,:]
    #         a[8*frame_div+j,:,:] = TR[i*frame_div + j,:,:]
        
    #     tifffile.imwrite(filepath + writepath + 'stack' + str(i) + '.tif', a)
```
